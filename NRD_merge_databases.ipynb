```python
#Function to calculate how long until their next readmission
def calculate_days_to_readmission(next_nrd, nrd_daystoevent, los):
    return next_nrd - nrd_daystoevent - los

#function to calculte if this admission is an index admission. Must fulfill criterias of not admission in december,not
#dying at this admission, nad have a valid length of stay
def calculate_index_admission(row):
    if (0 < row['dmonth'] < 12 and row['died'] == 0 and row['los'] >= 0):
        return 1
    else:
        return 0
    
#Calculate the number of total unique patients as well as number of total visits. then filter by age>18 and valid los
#The two print statements below will print out the # of unique patients (weighted), as well as the # of visits (weighted)

#print(core_cohort_2017.groupby('nrd_visitlink')['discwt'].mean().sum())
#print(core_cohort_2017['discwt'].sum())
core_cohort_2017 = core_cohort_2017[core_cohort_2017['age']>=18]
core_cohort_2017 = core_cohort_2017[core_cohort_2017['los']>=0]


#Read in hospital and severity to merge together
hospital_2017 = pd.read_csv('NRD_2017_Hospital.csv', usecols=['hosp_bedsize', 'h_contrl', 'hosp_nrd', 'hosp_urcat4', 'hosp_ur_teach'], dtype={'hosp_bedsize':'float16', 'h_contrl':'float16', 'hosp_nrd':'float64', 'hosp_urcat4':'float16', 'hosp_ur_teach':'float16',})
severity_2017 = pd.read_csv('NRD_2017_Severity.csv', dtype={'aprdrg':'float16', 'aprdrg_risk_mortality':'float16', 'aprdrg_severity':'float16', 'hosp_nrd':'float64', 'key_nrd':'float64'})

#Merge together dataframes and then sort them based off of their unique visitlink identifier, then their
#nrd_daystoevent that is ascending. This way all the visits are put in order
core_cohort_2017 = pd.merge(core_cohort_2017, hospital_2017, how='left', on='hosp_nrd')
print('merged hospital')
core_cohort_2017 = pd.merge(core_cohort_2017, severity_2017, how='left', on='key_nrd')
print('merged severity')
core_cohort_2017 = core_cohort_2017.sort_values(['nrd_visitlink', 'nrd_daystoevent'], ascending = (True, True))
print('sorted')

#Next_nrdday is the next time of admission for each patient. Days_to_readmission, index admission, and others are self explanatory
#Readmissino diagnosis and readmission los indicates the patient's *primary* diagnosis and los for the next time they are admitted.
#This is NA if the patient does not have a further readmission
core_cohort_2017['next_nrdday'] = core_cohort_2017.groupby('nrd_visitlink')['nrd_daystoevent'].shift(-1)
core_cohort_2017['days_to_readmission'] = calculate_days_to_readmission(core_cohort_2017['next_nrdday'], core_cohort_2017['nrd_daystoevent'], core_cohort_2017['los'])
core_cohort_2017['index_admission'] = core_cohort_2017.apply(calculate_index_admission, axis=1)
core_cohort_2017['readmission_diagnosis'] = core_cohort_2017.groupby('nrd_visitlink')['i10_dx1'].shift(-1)
core_cohort_2017['readmission_los'] = core_cohort_2017.groupby('nrd_visitlink')['los'].shift(-1)
```

```python
columns_to_add = ['key_nrd', 'hfrs_score', 'hfrs_severity', 'cc_score', 'cc_severity']

has_malignant_cancer_or_related_symptoms_count = 0
has_malignant_cancer_or_related_symptoms_ICD10 ={'C00', 'C01', 'C02', 'C03', 'C04', 'C05', 'C06', 'C07', 'C08', 'C09', 'C10', 'C11', 'C12', 'C13', 'C14',
                   'C15', 'C16', 'C17', 'C18', 'C19', 'C20', 'C21', 'C22', 'C25', 'C23', 'C24', 'C26', 'C32', 'C33', 'C34', 
                    'C30', 'C31', 'C37', 'C38', 'C39', 'C45', 'C40', 'C41', 'C48', 'C49', 'C43', 'C46', 'C84', 'C44', 'C50', 
                    'C53', 'C54', 'C55', 'C56', 'C51', 'C52', 'C57', 'C58', 'C61', 'C60', 'C62', 'C63', 'C67', 'C64', 'C65', 
                    'C66', 'C68', 'C69', 'C70', 'C71', 'C72', 'C47', 'C73', 'C74', 'C75', 'C82', 'C83', 'C84', 'C85', 'C86', 
                    'C81', 'C88', 'C90', 'C91', 'C92', 'C93', 'C94', 'C95', 'C96', 'C76', 'D46', 'D47', 'C7A', 'C77', 'C78', 
                    'C79', 'C80', 'C7B', 'D630', 'D6481', 'E340', 'E883', 'G131', 'G731', 'J910', 'K1233', 'K627', 
                    'M360', 'M361', 'M845', 'M906', 'O9A1', 'R180', 'R530', 'T66', 'Y632', 'Y842'}

hfrs10df = pd.read_csv('Frailty_ICD10_Codes.csv')
hfrs_dict = {}
for index, row in hfrs10df.iterrows():
    hfrs_dict[row['icd_10_codes']] = row['points']
    
#Charleson Comorbidity index categories
cc_MI = {'I21', 'I22', 'I252'}
cc_CHF = {'I1099', 'I110', 'I130', 'I132', 'I255', 'I420', 'I425', 'I426', 'I427', 'I428', 'I429', 'I43', 'I50', 'P290'}
cc_PVD = {'I70', 'I71', 'I731', 'I738', 'I739', 'I771', 'I790', 'I792', 'K551', 'K558', 'K559', 'Z958', 'Z959'}
cc_CVD = {'G45', 'G46', 'H340', 'I60', 'I61', 'I62', 'I63', 'I64', 'I65', 'I66', 'I67', 'I68', 'I69'}
cc_dementia = {'F00', 'F01', 'F02', 'F03', 'F051', 'G30', 'G311'}
cc_COPD = {'I278', 'I279', 'J40', 'J41', 'J42', 'J43', 'J44', 'J45', 'J46', 'J47', 'J60', 'J61', 'J62', 'J63', 'J64', 'J65', 'J66', 'J67', 'J684', 'J701', 'J703'}
cc_rheum = {'M05', 'M06', 'M315', 'M32', 'M33', 'M34', 'M351', 'M353', 'M360'}
cc_peptic_ulcer = {'K25', 'K26', 'K27'}
cc_mild_liver_disease = {'B18', 'K700', 'K701', 'K702', 'K703', 'K709', 'K713', 'K714', 'K715', 'K717', 'K73', 'K74', 'K760', 'K762', 'K763', 'K764', 'K768', 'K769', 'Z944'}
cc_diabetes_wo_complications = {'E100', 'E101', 'E106', 'E108', 'E109', 'E110', 'E111', 'E116', 'E118', 'E119', 'E120', 'E121', 'E126', 
                                'E128', 'E129', 'E130', 'E131', 'E136', 'E138', 'E139', 'E140', 'E141', 'E146', 'E148', 'E149'}
cc_diabetes_w_complications = {'E102', 'E103', 'E104', 'E105', 'E107', 'E112', 'E113', 'E114', 'E115', 'E117', 'E122', 'E123', 'E124', 
                               'E125', 'E127', 'E132', 'E135', 'E137', 'E142', 'E143', 'E144', 'E145', 'E147'}
cc_plegia = {'G041', 'G114', 'G801', 'G802', 'G81', 'G82', 'G830', 'G831', 'G832', 'G834', 'G839'}
cc_renal = {'I120', 'I131', 'N032', 'N033', 'N034', 'N035', 'N036', 'N037', 'N052', 'N053', 'N054', 'N055', 'N056', 'N057', 'N18', 'N19', 'N250', 'Z490', 'Z491', 'Z492', 'Z940', 'Z992'}
cc_malignancy = has_malignant_cancer_or_related_symptoms_ICD10
cc_severe_liver = {'I850', 'I859', 'I864', 'I982', 'K704', 'K711', 'K721', 'K729', 'K765', 'K766', 'K767'}
cc_metastasis = {'C77', 'C78', 'C79', 'C80'}
cc_AIDS = {'B20', 'B21', 'B22', 'B24'}

def calculate_hfrs_score(diagnosis):
    if diagnosis in hfrs_dict:
        return hfrs_dict[diagnosis]
    else:
        return 0

def calculate_cc_score(diagnosis, status_dict):
    if (str(icd_10_diagnoses)[0:3] in cc_MI or str(icd_10_diagnoses)[0:4] in cc_MI):
        status_dict['MI'] = True
        return 1
    if (str(icd_10_diagnoses)[0:3] in cc_CHF or str(icd_10_diagnoses)[0:4] in cc_CHF):
        status_dict['CHF'] = True
        return 1
    if (str(icd_10_diagnoses)[0:3] in cc_PVD or str(icd_10_diagnoses)[0:4] in cc_PVD):
        status_dict['PVD'] = True
        return 1
    if (str(icd_10_diagnoses)[0:3] in cc_CVD or str(icd_10_diagnoses)[0:4] in cc_CVD):
        status_dict['CVD'] = True
        return 1
    if (str(icd_10_diagnoses)[0:3] in cc_dementia or str(icd_10_diagnoses)[0:4] in cc_dementia):
        status_dict['dementia'] = True
        return 1
    if (str(icd_10_diagnoses)[0:3] in cc_COPD or str(icd_10_diagnoses)[0:4] in cc_COPD):
        status_dict['COPD'] = True
        return 1
    if (str(icd_10_diagnoses)[0:3] in cc_rheum or str(icd_10_diagnoses)[0:4] in cc_rheum):
        status_dict['rheum'] = True
        return 1
    if (str(icd_10_diagnoses)[0:3] in cc_peptic_ulcer or str(icd_10_diagnoses)[0:4] in cc_peptic_ulcer):
        status_dict['peptic_ulcer'] = True
        return 1
    if (str(icd_10_diagnoses)[0:3] in cc_mild_liver_disease or str(icd_10_diagnoses)[0:4] in cc_mild_liver_disease):
        status_dict['mild_liver'] = True
        return 1
    if (str(icd_10_diagnoses)[0:3] in cc_diabetes_wo_complications or str(icd_10_diagnoses)[0:4] in cc_diabetes_wo_complications):
        status_dict['diabetes_wo_complications'] = True
        return 1
    if (str(icd_10_diagnoses)[0:3] in cc_diabetes_w_complications or str(icd_10_diagnoses)[0:4] in cc_diabetes_w_complications):
        status_dict['diabetes_w_complications'] = True
        return 2
    if (str(icd_10_diagnoses)[0:3] in cc_plegia or str(icd_10_diagnoses)[0:4] in cc_plegia):
        status_dict['hemiplegia'] = True
        return 2
    if (str(icd_10_diagnoses)[0:3] in cc_renal or str(icd_10_diagnoses)[0:4] in cc_renal):
        status_dict['renal'] = True
        return 2
    if (str(icd_10_diagnoses)[0:3] in cc_malignancy or str(icd_10_diagnoses)[0:4] in cc_malignancy):
        status_dict['malignancy'] = True
        return 2
    if (str(icd_10_diagnoses)[0:3] in cc_severe_liver or str(icd_10_diagnoses)[0:4] in cc_severe_liver):
        status_dict['severe_liver'] = True
        return 3
    if (str(icd_10_diagnoses)[0:3] in cc_metastasis or str(icd_10_diagnoses)[0:4] in cc_metastasis):
        status_dict['metastasis'] = True
        return 6
    if (str(icd_10_diagnoses)[0:3] in cc_AIDS or str(icd_10_diagnoses)[0:4] in cc_AIDS):
        status_dict['AIDS'] = True
        return 6
    else:
        return 0
```

```python
total_patient_count = 0
frailty_data = []

for row in tqdm(core_cohort_2017.itertuples()):
    cc_status_dict = {'MI':False, 'CHF':False, 'PVD':False, 'CVD':False, 'dementia':False, 'COPD':False, 'rheum':False,
                     'peptic_ulcer':False, 'mild_liver':False, 'diabetes_wo_complications':False, 'diabetes_w_complications':False,
                     'hemiplegia':False, 'renal':False, 'malignancy':False, 'severe_liver':False, 'metastasis':False, 'AIDS':False}
    patient_frailties = [row.key_nrd,0,0,0,0]
    
    hfrs_score = 0
    cc_score = 0
    
    i=1
    while i<=40:
        #this allows us to iterate from i10_d1 - i10_d40
        column = 'i10_dx' + str(i)
        #Get the ICD10 code for that column
        icd_10_diagnoses = getattr(row, column)        
        #If the ddx is blank and it's not the first ddx (for some reason some of the first ddx is blank), then move on to the next row
        if icd_10_diagnoses is np.nan and i != 1:
            break
        
        hfrs_score += calculate_hfrs_score(str(icd_10_diagnoses)[0:3])
        cc_score += calculate_cc_score(icd_10_diagnoses, cc_status_dict)
        
        i+=1
    patient_frailties[1] = round(hfrs_score, 2)
    if hfrs_score >= 15:
        patient_frailties[2] = 2
    if 15 > hfrs_score >= 5:
        patient_frailties[2] = 1
    if hfrs_score < 5:
        patient_frailties[2] = 0
        
    patient_frailties[3] = cc_score
    if cc_score <3:
        patient_frailties[4] = 0
    if 5 > cc_score >= 3:
        patient_frailties[4] = 1
    if cc_score >= 5:
        patient_frailties[4] = 2
    

    frailty_data.append(patient_frailties)

# Load in morbidities data as pandas but will convert to dask later
frailty_df = pd.DataFrame(frailty_data, columns=columns_to_add)
#dask_comorbidities_df = dd.from_pandas(comorbidities_df, npartitions=12)

# dask merge is much more efficient if the keys are the index
frailty_df = frailty_df.set_index('key_nrd')

core_cohort_2017=core_cohort_2017.set_index('key_nrd')
```
